What is Unicode?

- Unicode is a standard for representing text from all the world‚Äôs writing systems.

- Every character (letter, number, symbol, emoji, etc.) is assigned a unique code point, which is just a number.


| Block Name         | Range           | What it contains                     |
| ------------------ | --------------- | ------------------------------------ |
| Basic Latin        | 0x0000‚Äì0x007F   | English letters, digits, punctuation |
| Latin-1 Supplement | 0x0080‚Äì0x00FF   | Accented letters, symbols            |
| Devanagari         | 0x0900‚Äì0x097F   | Hindi, Sanskrit characters           |
| Cyrillic           | 0x0400‚Äì0x04FF   | Russian, Bulgarian                   |
| Emoji              | 0x1F600‚Äì0x1F64F | Smiley faces, emojis      



Why Python can ‚Äúsee‚Äù them

- All modern languages (Python, Java, JavaScript, C#, etc.) support Unicode.
- Python has Unicode built-in.
- Strings store the code points internally.
- Functions like chr() and ord() let you go from number ‚Üî character.
- That‚Äôs why we can generate Hindi characters programmatically by iterating over code points.           |


```python
hindi_chars = []

for codepoint in range(0x0900, 0x0980):
  char = chr(codepoint)
  hindi_chars.append(char)

print(hindi_chars)
```

```python
english_chars = []

for codepoint in range(0x0000, 0x007F):
  char = chr(codepoint)
  english_chars.append(char)

print(english_chars)
```

```python
emojis = []

for codepoint in range(0x1F600, 0x1F64F):
  char = chr(codepoint)
  emojis.append(char)

print(emojis)
```

```python
vocab = set(hindi_chars) | set(english_chars) | set(emojis) | {' '}
print(vocab)
```

```python
vocab = list(vocab)
```

Add special tokens

Prepend these to your vocabulary:

- 'PAD' ‚Üí padding token

- 'UNK' ‚Üí unknown token

- 'SOS' ‚Üí start of sequence

- 'EOS' ‚Üí end of sequence

```python
vocab.extend(['<PAD>','<UNK>','<SOS>','<EOS>'])
```

```python
char2id = dict()
id2char = dict()

for index, char in enumerate(vocab):
  char2id[char] = index
  id2char[index] = char
  # print(id2char[char2id[char]] == char)

print(len(char2id)==len(id2char))
```

```python
import json

with open('char2id.json', 'w', encoding='utf-8') as f:
    json.dump(char2id, f, ensure_ascii=False, indent=2)

with open('id2char.json', 'w', encoding='utf-8') as f:
    json.dump(id2char, f, ensure_ascii=False, indent=2)
```

```python
import json

class Encoder():

  def __init__(self) -> None:
     # Load char2id
    with open('char2id.json', 'r', encoding='utf-8') as f:
        self.char2id = json.load(f)

    # Load id2char
    with open('id2char.json', 'r', encoding='utf-8') as f:
        self.id2char = json.load(f)

    self.id2char = {int(k): v for k, v in self.id2char.items()}

    self.unk_id = self.char2id['<UNK>']
    self.sos_id = self.char2id['<SOS>']
    self.eos_id = self.char2id['<EOS>']

  def encode(self,text):
    encoded = []

    encoded.append(self.sos_id)

    if type(text) != str:
      text = str(text)

    else:
      for char in text:
        if char in self.char2id:
          token_id = self.char2id[char]
          encoded.append(token_id)

        else:
          token_id = self.unk_id
          encoded.append(token_id)

    encoded.append(self.eos_id)
    return encoded


  def decode(self,ids):
    chars = []

    for token_id in ids:
      if token_id in [self.sos_id, self.eos_id]:
        continue
      if token_id in self.id2char:
        chars.append(self.id2char[token_id])
      else:
        chars.append('<UNK>')

    decoded = ''.join(chars)
    return decoded
```

```python
obj = Encoder()
```
